```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df_train = pd.read_pickle('/Users/aravindrajeshmenon/Documents/DataScienceProjects/Projects/Fraud-Detection/data/df_train_cleaned.pkl')
```

```python
plt.figure(figsize = (8,8))
sns.countplot(data = df_train, x = 'isFraud')
plt.show()
```

```python
df_train.groupby(df_train['isFraud'])['TransactionAmt'].mean()
```

```python
plt.figure(figsize=(10,6))
df_train.boxplot(column='TransactionAmt', by='isFraud')
plt.title('Transaction Amount by Fraud')
plt.show()
```

```python
numeric_cols = df_train.select_dtypes(include=['number']).columns
```

```python
top_cols = df_train[numeric_cols].corr()['isFraud'].abs().sort_values(ascending=False)[:10].index
plt.figure(figsize = (10,10))
sns.heatmap(df_train[top_cols].corr(), annot=True, cmap='coolwarm')
plt.show()
```
