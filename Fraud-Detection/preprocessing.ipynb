Since both the data frames have transaction ID's, we can concatenate both the tables into a single table.

```python
df_train = pd.merge(left = train_transaction, right = train_identity, on = 'TransactionID', how = 'left')
df_test = pd.merge(left = test_transaction, right = test_identity, on = 'TransactionID', how = 'left')
```

```python
missing_values = df_train.isna().sum().reset_index(name = 'count')
```

```python
cols = list(df_train.columns)
print('Column name', "-----------------", 'Null values')
columns = 0 
for col in cols:
    if df_train[col].isna().sum() > 0: 
        columns += 1 
        print(f'{col}: ----------------- {df_train[col].isna().sum()}')
print(f'Total number of columns with missing value = {columns}')    
```

We see a lot of columns have missing values with 412 columns in total having missing values. Now let us see how many columns have a high degree of missing values, to the point where imputation of any sort might not be precise and may lead to compromising the data quality. 

```python
high_missing_values = {}
for col in cols: 
    if df_train[col].isna().sum() > 0:
        prop = round(df_train[col].isna().sum()/len(df_train), 3)
        if prop > 0.5:
            print(f'{col} --------------- {prop}') 
            high_missing_values[col] = prop
            
```

Now, we see that certain columns have extremely high number of missing values with some even having extremely high proportions rising to about 99%. These columns can be removed as they provide no real information and at most serves as a source of noise in the system. I've chosen 80% as an arbitrary threshold for removing columns

```python
cols_to_drop = []
for col in high_missing_values.keys():
    if high_missing_values[col] > 0.8:
        cols_to_drop.append(col)


```

```python
df_train = df_train.drop(columns= cols_to_drop)
df_test = df_test.drop(columns = cols_to_drop)
```

```python
df_train.info()
```

```python
float_cols = df_test.select_dtypes(include = 'float64').columns
df_train[float_cols] = df_train[float_cols].astype('float32')
df_test[float_cols] = df_test[float_cols].astype('float32')

int_cols = df_test.select_dtypes(include = 'int64').columns
df_train[int_cols] = df_train[int_cols].astype('int32')
df_test[int_cols] = df_test[int_cols].astype('int32')

print(df_train.info(memory_usage='deep'))
```

```python
df_train.to_pickle('data/df_train_cleaned.pkl')
df_test.to_pickle('data/df_test_cleaned.pkl')
```
